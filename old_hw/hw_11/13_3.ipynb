```python
import matplotlib.pyplot as plt
```

```python
# calculate p(x) using coefficients per formula in (a)
def getP_a(x):
    p=x**9
    p-=18*(x**8)
    p+=144*(x**7)
    p-=672*(x**6)
    p+=2016*(x**5)
    p-=4032*(x**4)
    p+=5376*(x**3)
    p-=4608*(x**2)
    p+=2304*x
    p-=512
    return(p)

# calculate p(x) using (x-2)^9 per part (b)
def getP_b(x):
    p=(x-2)**9
    return(p)
```

```python
# calculate results
x_list = []
p_a = []
p_b = []
for i in range(1920,2080):
    x = i/1000 # range only uses integers so divide to get desired float
    x_list.append(x)
    p_a.append(getP_a(x))
    p_b.append(getP_b(x))
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1, 2, 1) # row 1, col 2 index 1
plt.plot(x_list, p_a)
plt.title("Results from (a)")
plt.xlabel('x ')
plt.ylabel('p(x) ')

plt.subplot(1, 2, 2) # index 2
plt.plot(x_list, p_b)
plt.title("Results from (b)")
plt.xlabel('x ')
plt.ylabel('p(x) ')

plt.show()
```
