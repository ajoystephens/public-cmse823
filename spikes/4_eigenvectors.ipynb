```python
# practical QR gives all eigenvalues, but does not give all eigenvectors

def this_practical_qr_algorithm(matrix,k_max,zero_cutoff):
    m = matrix.shape[0]
    A = np.ones((k_max,m,m),dtype=float)
    A[0],Q_0 = fullHessenberg(matrix)
    shift = np.ones(k_max)
    # eig=np.zeros(m,dtype=np.double)
    eig=[]
    eigenvect = Q_0
    
    # print(matrix.round(5))
    for k in range(1,k_max):
        # print('k:',k)
        shift[k] = A[k-1,m-1,m-1]
        if m==1: 
            # print('last:',shift[k])
            # return([shift[k].round(10)])
            return([shift[k]])
        Q,R = householder(A[k-1,:,:]-shift[k]*np.identity(m))
        A[k,:,:] = np.matmul(R,Q)+shift[k]*np.identity(m)
        eigenvect = np.matmul(eigenvect,Q)
        # if not np.isnan(eigenvect).any():
            # print('eigenvect')
            # print(eigenvect)
        # print(A[k,:,:].round(5))
        for j in range(m-1):
            # print('  j:',j,'val:',A[k,j,j+1])
            if np.abs(A[k,j,j+1]) < zero_cutoff:
                # print('shift:',shift[k],',',A[k,j,j+1])
                A[k,j,j+1]=0
                A[k,j+1,j]=0
                # print('shift:',shift[k],',',A[k,j,j+1])
                # break
                A_1 = A[k,:j+1,:j+1]
                A_2 = A[k,j+1:,j+1:]
                # eig=eig+[shift[k].round(8)]
                if A_1.shape[0]>0: 
                    e_temp = practical_qr_algorithm(A_1,k_max,zero_cutoff)
                    # print('eig:',eig,'etemp:',e_temp)
                    if np.abs(e_temp)[0]>0:eig=eig+e_temp
                    # if np.abs(e_temp)[0]>0:eig=np.concatenate(eig,e_temp)
                if A_2.shape[0]>0: 
                    e_temp = practical_qr_algorithm(A_2,k_max,zero_cutoff)
                    # print('eig:',eig,'etemp:',e_temp)
                    if np.abs(e_temp)[0]>0:eig=eig+e_temp
                    # if np.abs(e_temp)[0]>0:eig=np.concatenate(eig,e_temp)
                # return(eig)

    # print('not found')  
    # return(shift[k])
    return(eig)
```

```python
A = np.array(
    [[1,2,3,4],
    [2,1,2,2],
    [3,2,1,3],
    [4,2,3,1]]
)

# A[0,0:A.shape[0]]
eigvals = this_practical_qr_algorithm(A,20,1e-5)
print(eigvals)

print()
print('NP Results')
np_results = np.linalg.eig(A)
print(np_results[0])
print(np_results[1])

# j=2
# print(A[:j,:j])
# print(A[j:,j:])
```

```python
# gets the eigenvector corresponding to some eigenvalue
#  uses one itteration of Alg 27.2 from book, pdf pg 220, book pg 206
def getEigenvector(matrix, eigenvalue):
    A = matrix.copy().astype(np.double)
    n = matrix.shape[0]
    v = _getSomeUnitVector(n)
    w = solveSystem(A-eigenvalue*np.identity(n),v)
    eigenvector = w/np.linalg.norm(w,2)
    return(eigenvector)
    

def _getSomeUnitVector(n):
    v = np.random.randint(0,10,n)
    v = v/np.linalg.norm(v,2)
    return(v)
```
